```python
import numpy as np
from Equations import System
from scipy.optimize import fmin
from matplotlib import pyplot as plt
```

```python
# solving for q
sys = System(u_inf=1,D=1)
q_init = np.random.rand(5,1)
q_solved = fmin(func = sys.system_constraint_error,x0 = q_init)
q_solved
```

```python
# plotting and visualization
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stream Function = 0')

# stream function
n = 100
Z = np.zeros([n,n])
x_interval = np.linspace(-5,3,n)
y_interval =np.linspace(-1,1,n)

for j, y in enumerate(y_interval):
    for i, x in enumerate(x_interval):
        Z[j,i] = sys.streamfunction(x,y,q_solved)

X, Y = np.meshgrid(x_interval,y_interval)
contour = plt.contour(X,Y,Z, levels = 0)

# source and sinks
plt.scatter([-3,-2,-1,0,1],[0,0,0,0,0], label = 'Sources/Sinks')

plt.legend()
plt.show()
```
